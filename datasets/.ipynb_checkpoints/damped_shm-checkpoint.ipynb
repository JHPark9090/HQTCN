# Step 1: Import all necessary libraries

```python
import numpy as np
from scipy.integrate import odeint
import math
import matplotlib.pyplot as plt
import torch
from torch.autograd import Variable # Import the deprecated Variable for accuracy
from sklearn.preprocessing import MinMaxScaler

print("--- Step 1: Libraries Imported ---")
print("We need numpy for array manipulation, scipy.integrate.odeint to solve the differential equation,")
print("matplotlib for plotting, torch to create tensors for the model, and sklearn for data normalization.\n")
```

# Step 2: Define and Simulate the Physical System (Damped Pendulum)


```python
print("--- Step 2: Simulating the Damped Pendulum ---")

# This function defines the system of differential equations for the pendulum.
# It's the mathematical model of the pendulum's motion.
def system(theta, t, b, g, l, m):
    theta1 = theta[0]  # The angle of the pendulum
    theta2 = theta[1]  # The angular velocity of the pendulum
    dtheta1_dt = theta2
    dtheta2_dt = -(b / m) * theta2 - (g / l) * math.sin(theta1)
    dtheta_dt = [dtheta1_dt, dtheta2_dt]
    return dtheta_dt

# Set the physical constants for the simulation
b = 0.15  # Damping factor
g = 9.81  # Gravity
l = 1.0   # Length of the pendulum
m = 1.0   # Mass of the bob

# Set the initial conditions: [initial angle, initial angular velocity]
# The pendulum starts at angle 0 but is given an initial "push".
theta_0 = [0, 3]

# Create an array of 240 time points from t=0 to t=20
t = np.linspace(0, 20, 240)

# This is the main simulation step. `odeint` solves the `system` of equations.
# The output `theta` will be a 240x2 array where:
# Column 0 is the angle (theta1) at each time step.
# Column 1 is the angular velocity (theta2) at each time step.
theta = odeint(system, theta_0, t, args=(b, g, l, m))

print("Simulation complete. We now have the raw angle and angular velocity data.")
print("Time dimension: 20 seconds, 240 time points.")
print("In the QLSTM paper, the angular velocity is predicted.\n")

# Visualize the raw simulation output
plt.figure(figsize=(12, 5))
plt.plot(t, theta[:, 0], 'b-', label='Angle (theta1)')
plt.plot(t, theta[:, 1], 'r--', label='Angular Velocity (theta2)')
plt.title('Raw Simulation Output of the Damped Pendulum')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

# Step 3: Normalize the Data


```python
print("\n--- Step 3: Normalizing the Data for the Model ---")
print("The QLSTM model will learn to predict the angular velocity (the red dashed line above).")
print("Neural networks train best when input data is scaled to a small range, like [-1, 1].")

# Isolate the angular velocity data
# Reshape to a 2D array (n_samples, n_features) for scaling
# (240,) -> (240, 1)
angular_velocity_raw = theta[:, 1].reshape(-1, 1)   

# Use scikit-learn's MinMaxScaler to scale the data to the range [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
dataset = scaler.fit_transform(angular_velocity_raw)

print("\nData has been normalized.\n")

# Visualize the normalized dataset
plt.figure(figsize=(12, 5))
plt.plot(t, dataset, 'g-', label='Normalized Angular Velocity')
plt.title('Normalized Dataset for QLSTM Training')
plt.xlabel('Time')
plt.ylabel('Normalized Value')
plt.ylim(-1.1, 1.1)
plt.legend()
plt.grid(True)
plt.show()
```

# Step 4: Transform Data into Sequences for the LSTM


```python
print("\n--- Step 4: Creating Input Sequences and Target Values ---")
print("LSTMs learn from sequences. We need to transform our flat time-series into (input, output) pairs.")
print("The task is: given a sequence of past values, predict the next value.")

def transform_data_single_predict(data, seq_length):
    """
    Converts a flat time-series into input sequences (x) and target values (y).
    This uses a "sliding window" approach.
    """
    x = []
    y = []
    
    for i in range(len(data) - seq_length - 1):  # 240 time points -> 240-4-1 = 235 examples
        # Define the sliding window
        _x = data[i:(i + seq_length)]
        _y = data[i + seq_length]
        x.append(_x)
        y.append(_y)
        
    # 1. np.array(x) creates a 3D array of shape (num_sequences, seq_length, 1)
    # 2. .reshape(-1, seq_length) flattens it into a 2D array of shape (num_sequences, seq_length)
    #    This is done because the main QLSTM_v0.py script expects a 2D tensor.
    x_np_reshaped = np.array(x).reshape(-1, seq_length)
    y_np = np.array(y)
    
    # The original script uses a deprecated Variable wrapper, we'll use torch.tensor directly
    x_var = Variable(torch.from_numpy(x_np_reshaped).float())
    y_var = Variable(torch.from_numpy(y_np).float())
    
    return x_var, y_var

# Set the sequence length (how many past steps the model sees)
# The original script uses a default of 4.
seq_len = 4

# Transform the normalized data
x, y = transform_data_single_predict(dataset, seq_len)

print(f"Using a sequence length of {seq_len}, we created {len(x)} training examples.")
print("Let's look at the first few examples to understand the data format.\n")

# Visualize the 2D input tensor
print("The input tensor 'x' is now 2D, as created by the original script.")
for i in range(3):
    print(f"Example {i+1}:")
    print(f"  Input Sequence (x): {x[i].numpy().round(3)}")
    print(f"  Target Value (y):   {y[i].numpy().round(3)}\n")
```

# Step 5: Final Check of Data Shapes

```python

print("\n--- Step 5: Final Data Shape for the QLSTM Model ---")

print("The data-loading script produces a 2D input tensor 'x'.")
print(f"Shape of the input sequences tensor (x): {x.shape}")
print(f"Shape of the target values tensor (y):  {y.shape}")

print("\nThe main QLSTM_v0.py script then adds the 3rd dimension back using '.unsqueeze(2)'.")
x_transformed = x.unsqueeze(2)
print(f"Shape after '.unsqueeze(2)' in the main script: {x_transformed.shape}")
print("\nThis final 3D shape is what the QLSTM model actually receives for training.")
```
